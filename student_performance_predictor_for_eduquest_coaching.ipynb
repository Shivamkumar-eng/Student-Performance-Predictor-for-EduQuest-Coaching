```python
# import library
import pandas as pd
```

```python
# Read dataset
df= pd.read_csv('/content/Student Performance Predictor for EduQuest Coaching.csv')
```

```python
df.info()
```

# In df.info() only  extra_curricular_involvement  has  628 value
## which means is contain missing values so we will ignore this

```python
df= df.dropna()
```

# throug df.dropna() we  drop the  misssing value

# NOW WE DESCRIBE THE DATA THROUGH df.describe()
# which ignore the non numeric or object data

```python
df.describe()
```

# THROUGH df.describe(include='all') we force to include all

```python
df.describe(include='all')
```

# //ENCODING

```python
# Here this syntax give the catagories of data
df['gender'].value_counts()
```

**NOW WE  GOING TO REPLACE THIS INTO 0 AND 1**



```python
df['gender']= df['gender'].replace({'Male':0, 'Female':1})
```

```python
df['gender'].value_counts()
```

```python
df['parental_education'].value_counts()
```

**NOW WE CONVERT  parental_education strings INTO INTIGER FORM INTO A SERIAL MANER**

```python
df['parental_education']=df['parental_education'].replace({'High School':0,'Diploma':1,'Graduate':2,'Postgraduate':3})
```

```python
df['internet_access']=df['internet_access'].replace({'Yes':1,'No':0})
```

```python
df.info()
```

# extra_curricular_involvement and tutor_support are also  object so convert this in numeric for as others

```python
df['extra_curricular_involvement'].value_counts()
```

```python
df['extra_curricular_involvement']= df['extra_curricular_involvement'].replace({'Low':0,'Moderate':1,'High':2})
```

```python
df['tutor_support'].value_counts()
```

**now we going to check again**



```python
df['tutor_support']=df['tutor_support'].replace({'Yes':1,'No':0})
```

```python
df.info()
```

***// DEFINE X AND Y ***

```python
df.columns
```

```python
y=df['final_exam_score']
x=df[['gender', 'age', 'parental_education', 'family_income',
       'internet_access', 'previous_exam_score', 'attendance_rate',
       'homework_completion_rate', 'class_participation_score',
       'number_of_absences', 'extra_curricular_involvement',
       'learning_hours_per_week', 'tutor_support',]]
```

**NOW WE GOING TO SPILIT THE DATA INTO TRAIN AND TEST**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=2529)
```

**//NOW WE GOING TO SELECT THE MODEL**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

**//TRAIN THE MODEL**

```python
model.fit(x_train,y_train)
```

WE USING LINER REGRESSTION AND THE FORMULA OF LINER REGRESSION IS
 Y= A+BX +E(WHERE "E" IS ERROR )
# THE LINE EQUATION IS Y= A+BX





```python
model.intercept_,model.coef_
```

**//PREDICTION**

```python
y_pred = model.predict(x_test)
```

**NOW THE FINAL STEP IS EVALUATE**

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

#  -> THE MAPE( mean absolute percentage error) for this problem using LinearRegression is 5.71%
